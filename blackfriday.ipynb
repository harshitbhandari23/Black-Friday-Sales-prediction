# creating pie chart

```python
d=df["Gender"].value_counts().reset_index()
plt.figure(figsize=(10,5))
plt.title("no of males and females")
plt.pie(d["count"],labels=d["Gender"],autopct="%0.1f",wedgeprops={"edgecolor":"black"})
plt.legend()
plt.show()
```

# pie cart for the same

```python
a=df.groupby(by="Gender")["Purchase"].mean().reset_index()
plt.figure(figsize=(10,5))
plt.title("average purchases")
plt.pie(a["Purchase"],labels=a["Gender"],autopct="%0.1f")
plt.show()
```

# combing gender and marital status

```python
l=[]
for i,row in df.iterrows():
    l.append(row["Gender"]+"_"+str(row["Marital_Status"]))
df["gender_status"]=l
```

```python
snp.countplot(data=df,x="gender_status",palette="pastel",hue="City_Category")
```

```python
a=df.groupby(by="gender_status")["Purchase"].sum().reset_index()
snp.barplot(data=a,x="gender_status",y="Purchase",palette="pastel")
```
